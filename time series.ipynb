```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
```

```python
sales=[100,200,300,400,500,600,700]
feature=3
x,y=[],[]
for i in range(len(sales)-feature):
    x.append(sales[i:i+feature])
    y.append(sales[i+feature])

    
```

```python
x=np.array(x)
y=np.array(y)
print(x,y)
x.shape
```

```python
x=x.reshape(x.shape[0],1,feature)
x
```

```python
model=Sequential()
model.add(LSTM(50,activation='relu',input_shape=(1,feature)))
```

```python
model.add(Dense(1))
```

```python
model.compile(loss='mse',optimizer='adam')
```

```python
model.fit(x,y,epochs=2000,verbose=1)
```

```python
seq_data=sales[-feature:]
seq_data=np.array(seq_data).reshape(1,1,feature)
model.predict(seq_data,verbose=0)[0][0]
```
